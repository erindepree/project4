```python
month_temp = pd.DataFrame(columns=[
    'month', 
    'avg_mean', 'avg_std', 'avg_extreme_max', 'avg_extreme_min',
    'max_mean', 'max_std', 'max_extreme', 'days_above_max_extreme',
    'min_mean', 'min_std', 'min_extreme', 'days_below_min_extreme'
])
```

```python
for mm in range(1,13):
    # mean temperatures
    avg_mean = data[data['month']==mm]['temp_mean'].mean()
    avg_std = data[data['month']==mm]['temp_mean'].std()
    
    
    # heat wave
    max_mean = data[data['month']==mm]['temp_max'].mean()
    max_std = data[data['month']==mm]['temp_max'].std()
    max_extreme = max_mean + 2.5*max_std
    
    #month_temp.loc[0,'max_extreme'] = month_temp.loc[0,'max_mean'] + 2.5*month_temp.loc[0,'max_std']
    #month_temp['days_above_max_extreme'] = data[
    #    (data['month']==mm) & (data['temp_max'] > month_temp['max_extreme'])
    #].count()
    
    # cold waves
    min_mean = data[data['month']==mm]['temp_min'].mean()
    min_std = data[data['month']==mm]['temp_min'].std()
    min_extreme = min_mean - 2.5*min_std
    #month_temp.loc[0,'min_extreme'] = month_temp.loc[0,'max_mean'] - 2.5*month_temp.loc[0,'max_std']
    #month_temp['days_below_min_extreme'] = data[
    #    (data['month']==mm) & (data['temp_min'] < month_temp['min_extreme'])
    #].count()
    
    month_temp.loc[mm] = {
        'month': mm, 
        'avg_mean': avg_mean,
        'avg_std': avg_std,
        'avg_extreme_max': avg_mean + 2.5*avg_std, 
        'avg_extreme_min': avg_mean - 2.5*avg_std,
        'max_mean': max_mean, 
        'max_std': max_std, 
        'max_extreme': max_extreme, 
        'days_above_max_extreme': len(data[(data['month']==mm) & (data['temp_max'] >= max_extreme)]),
        'min_mean': min_mean, 
        'min_std': min_std, 
        'min_extreme': min_extreme, 
        'days_below_min_extreme': len(data[(data['month']==mm) & (data['temp_min'] <= min_extreme)])
    }
```

```python
extreme_df = pd.DataFrame(columns=[
    'is_above_max_extreme', 'is_below_min_extreme'
])
```

```python
for i in data.index:
    is_extreme_heat = False
    is_extreme_cold = False
    month = data.loc[i, 'date'].month

    # check if above or below extreme temperatures
    if data.loc[i, 'temp_max'] >= month_summary.loc[month, 'max_extreme']:
        is_extreme_heat = True
    elif data.loc[i, 'temp_min'] <= month_summary.loc[month, 'min_extreme']:
        is_extreme_cold = True

    extreme_df.loc[i] = {
        'is_above_max_extreme': is_extreme_heat, 
        'is_below_min_extreme': is_extreme_cold
    }
```

```python
extreme_data = pd.merge(left= data, right= extreme_df, left_index=True, right_index=True)
```

```python
temp_waves = pd.DataFrame(columns=['is_heat_wave', 'is_cold_wave'])
```

```python
for i in extreme_data.index:
    date = extreme_data['date']

    is_heat_wave = False
    is_cold_wave = False
    
    if extreme_data.loc[i, 'is_above_max_extreme'] == True and (
        extreme_data.loc[i-1, 'is_above_max_extreme'] == True 
        or extreme_data.loc[i+1, 'is_above_max_extreme'] == True
    ):
        is_heat_wave = True

    if extreme_data.loc[i, 'is_below_min_extreme'] == True and (
        extreme_data.loc[i-1, 'is_below_min_extreme'] == True
        or extreme_data.loc[i+1, 'is_below_min_extreme'] == True
    ):
        is_cold_wave = True

    temp_waves.loc[i] = {'is_heat_wave': is_heat_wave, 'is_cold_wave': is_cold_wave}
```

```python
final_data = pd.merge(
    left = extreme_data[[
        'date', 'temp_max', 'temp_min', 'latitude', 'longitude', 
        'is_above_max_extreme', 'is_below_min_extreme'
    ]], 
    right= temp_waves, 
    left_index= True, 
    right_index= True
)
```
