```python
def extreme_temperature(month, file_name):
    mon_sum = pd.read_csv('../data/temp_events/' + file_name).drop(columns='Unnamed: 0').set_index('month')

    return mon_sum.loc[month, 'max_extreme'], mon_sum.loc[month, 'min_extreme']
```

```python
df = pd.DataFrame(columns = ['extreme_max', 'extreme_min'])

for i in data.index:
    month = datetime.date(datetime.fromisoformat(data.loc[i,'date'])).month
    lat = int(data.loc[i,'latitude'])
    long = int(data.loc[i,'longitude'])
    
    if f"monthly_summary_lat={lat}_long={long}.csv" in file_list:
        extreme_max, extreme_min = extreme_temperature(month, f"monthly_summary_lat={lat}_long={long}.csv")
    elif f"monthly_summary_lat={lat-1}_long={long}.csv" in file_list:
        extreme_max, extreme_min = extreme_temperature(month, f"monthly_summary_lat={lat}_long={long}.csv")
    elif f"monthly_summary_lat={lat}_long={long-1}.csv" in file_list:
        extreme_max, extreme_min = extreme_temperature(month, f"monthly_summary_lat={lat}_long={long}.csv")
    elif f"monthly_summary_lat={lat-1}_long={long-1}.csv" in file_list:
        extreme_max, extreme_min = extreme_temperature(month, f"monthly_summary_lat={lat}_long={long}.csv")
    else:
        print(f"Can't find the monthly summary file for lat={lat} and long={long}")
        

        
    df.loc[i,'extreme_max'] = extreme_max
    df.loc[i,'extreme_min'] = extreme_min
```
