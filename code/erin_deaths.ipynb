```python
df = pd.read_excel('../data/raw/emdat_extreme_temperature_deaths_2000_2025.xlsx')

df
```

```python
df1 = df.drop(columns=[
    'Historic', 'Classification Key', 'Disaster Group', 'Disaster Subgroup', 'Disaster Type', 
    'External IDs', 'Event Name', 'ISO','OFDA/BHA Response', 'Appeal', 'Declaration', 
    "AID Contribution ('000 US$)", 'River Basin','No. Injured', 'No. Affected', 'No. Homeless', 
    'Total Affected', "Reconstruction Costs ('000 US$)", "Reconstruction Costs, Adjusted ('000 US$)",
    "Insured Damage ('000 US$)", "Insured Damage, Adjusted ('000 US$)", "Total Damage ('000 US$)", 
    "Total Damage, Adjusted ('000 US$)", 'CPI', 'Admin Units'
])
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(subset=['Total Deaths'])
```

```python
df1 = df1.dropna(subset=['Start Day', 'End Day'])
```

```python
df1 = df1.reset_index(drop=True)
```

```python
df1.info()
```

```python
df1[['Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day']] = df1[[
    'Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day'
]].astype('int64')
```

```python
for i in df1.index:
    df1.loc[i,'start_date'] = date(
        df1.loc[i,'Start Year'], 
        df1.loc[i,'Start Month'], 
        df1.loc[i,'Start Day']
    )
    df1.loc[i,'end_date'] = date(
        df1.loc[i,'End Year'], 
        df1.loc[i,'End Month'], 
        df1.loc[i,'End Day']
    )
    df1.loc[i, 'duration_days'] = df1.loc[i,'end_date'] - df1.loc[i,'start_date']
```

```python
df1.loc[162,'duration_days']
```

```python
df2 = df1.drop(columns= ['Start Day', 'End Year', 'End Month', 'End Day', 'Latitude', 'Longitude', 
                         'Origin', 'Associated Types', 'Entry Date', 'Last Update', 'DisNo.'])
```

```python
df2
```

```python
df3= df2[df2['duration_days']>timedelta(days=1)]
```

```python
cold_deaths_reported = df3[df3['Disaster Subtype'] == 'Cold wave'].reset_index(drop=True)
```

```python
cold_deaths_reported
```

```python
heat_deaths_reported = df3[df3['Disaster Subtype'] == 'Heat wave'].reset_index(drop=True)
```
