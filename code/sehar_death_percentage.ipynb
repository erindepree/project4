```python
# Create new df with temp, location, and death related columns
selected_columns_order = ['Disaster Group','Disaster Subgroup','Disaster Type','Disaster Subtype',
                          'Region','Subregion','Country','Location','Origin','Latitude','Longitude',
                          'Associated Types', 'Magnitude', 'Magnitude Scale',
                          'Total Deaths', 'Total Affected',
                          'Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day']

pd.set_option('display.max_rows', None)
temp_death_df = og_temp[selected_columns_order]
temp_death_df.head()
```

```python
print(temp_death_df.shape)
temp_death_df.isnull().sum()
```

```python
# reduce features even more
selected_columns_order2 = ['Disaster Subgroup','Disaster Type','Disaster Subtype',
                          'Region','Subregion','Country',
                          'Total Deaths', 'Total Affected',
                          'Start Year', 'Start Month', 'End Year', 'End Month']

temp_death2_df = og_temp[selected_columns_order2]
temp_death2_df.head()
```

```python
# Filter for years 2020-2024 inclusive
df_filtered = temp_death2_df[temp_death2_df['Start Year'].between(2020, 2024)]

# Define the grouping columns in order
group_cols = ['Disaster Subgroup', 'Disaster Type', 'Disaster Subtype',
              'Region', 'Subregion', 'Country', 'Start Year']

# Group by above order
grouped = df_filtered.groupby(group_cols)

# Aggregations: sum deaths and affected counts
result = grouped.agg({
    'Total Deaths': 'sum',
    'Total Affected': 'sum'
}).reset_index()

result
```

##### Want percentages of deaths due to extreme temperatures for each country for each year 

```python
# Define disaster subtypes
heat_wave = 'Heat wave'
cold_wave_types = ['Cold wave', 'Severe winter conditions']

# Group by country and year, aggregate
agg_df = (
    df_filtered.groupby(['Country', 'Start Year'])
    .apply(lambda x: pd.Series({
        'Total Death Heat Wave': x.loc[x['Disaster Subtype'] == heat_wave, 'Total Deaths'].sum(),
        'Total Death Cold Wave + Severe Winter Conditions': x.loc[x['Disaster Subtype'].isin(cold_wave_types), 'Total Deaths'].sum(),
        'Total Death Country Year': x['Total Deaths'].sum()
    }))
    .reset_index()
)

# Calculate percentages
agg_df['Pct Death Heat Wave'] = (agg_df['Total Death Heat Wave'] / agg_df['Total Death Country Year'] * 100).fillna(0)
agg_df['Pct Death Cold Wave + Severe Winter Conditions'] = (
    agg_df['Total Death Cold Wave + Severe Winter Conditions'] / agg_df['Total Death Country Year'] * 100
).fillna(0)

agg_df
```

##### Let's see which countries are reporting out of the 5 years

```python
print("Total Number of Countries Reporting 2020-2024:\n", agg_df['Country'].nunique(),"Countries")
agg_df['Country'].value_counts()
```

##### There are ~192 countries in the world so almost all are reporting now...

```python
# How many for each frequency 
# Get the frequency of each country (how many rows each country has)
country_counts = agg_df['Country'].value_counts()

# Then count how many countries have frequency 5, 4, 3, 2, and 1
frequency_counts = country_counts.value_counts()

# Select the counts of interest (5,4,3,2,1) and fill 0 if missing
result = frequency_counts.reindex([5,4,3,2,1], fill_value=0)

result
```
